```python
from PIL import Image
import numpy as np
from PIL import ImageChops
import matplotlib.pyplot as plt
from PIL import ImageStat as stat

obraz1 = Image.open('obraz1.png')
print(f'Tryb obrazu: {obraz1.mode}')
```

```python
r, g, b, a = obraz1.split()

obraz3 = Image.new('RGB', obraz1.size, (255,255,255))
obraz3.paste(obraz1, (0, 0), a)
print(f'Tryb obrazu: {obraz3.mode}')
```

```python
# tryby =  ['CMYK','YCbCr','HSV']
# # plt.figure(figsize=(16, 16))
# i=1
# for t in tryby:
#     file_name = "tryb_"+ str(t)
#     im_c = obraz3.convert(t)

#     plt.subplot(1, 3, i)
#     plt.title(str(file_name))
#     plt.imshow(im_c, "gray")
#     plt.axis('off')
#     i +=1

#     plt.tight_layout()
#     plt.savefig('fig1.png')
tryby =  ['CMYK','YCbCr','HSV']

for i, t in enumerate(tryby, start=1):
    file_name = "tryb_"+ str(t)
    im_c = obraz3.convert(t)

    plt.subplot(1, 3, i)
    plt.title(str(file_name))
    plt.imshow(im_c, "gray")
    plt.axis('off')
    

    plt.tight_layout()
    plt.savefig('fig1.png')
```

```python
tryby =  ['CMYK','YCbCr','HSV']
file_name = "tryb_"+ str(tryby[0])
im_c = obraz3.convert(tryby[0])
c, m, y, k = im_c.split()

plt.subplot(1, 4, 1)
plt.title(f'{file_name}:\nkanał C')
plt.imshow(c, "gray")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title(f'{file_name}:\nkanał M')
plt.imshow(m, "gray")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title(f'{file_name}:\nkanał Y')
plt.imshow(y, "gray")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title(f'{file_name}:\nkanał K')
plt.imshow(k, "gray")
plt.axis('off')

plt.tight_layout()
plt.savefig('cmyk.png')
```

```python
tryby =  ['CMYK','YCbCr','HSV']
file_name = "tryb_"+ str(tryby[1])
im_c = obraz3.convert(tryby[1])
Y, Cb, Cr = im_c.split()

plt.subplot(1, 3, 1)
plt.title(f'{file_name}:\nkanał Y')
plt.imshow(Y, "gray")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title(f'{file_name}:\nkanał Cb')
plt.imshow(Cb, "gray")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title(f'{file_name}:\nkanał Cr')
plt.imshow(Cr, "gray")
plt.axis('off')

plt.tight_layout()
plt.savefig('ycbcr.png')
```

```python
tryby =  ['CMYK','YCbCr','HSV']
file_name = "tryb_"+ str(tryby[2])
im_c = obraz3.convert(tryby[2])
h, s, v = im_c.split()

plt.subplot(1, 3, 1)
plt.title(f'{file_name}:\nkanał H')
plt.imshow(h, "gray")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title(f'{file_name}:\nkanał S')
plt.imshow(s, "gray")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title(f'{file_name}:\nkanał V')
plt.imshow(v, "gray")
plt.axis('off')

plt.tight_layout()
plt.savefig('hsv.png')
```

```python
obraz4 = Image.open('obraz4.jpg')
print(f'Przed zmianą:\nWymiary obrazu: {obraz1.size},\nwymiary tła: {obraz4.size}.', end='\n\n')
obraz4 = obraz4.resize(obraz1.size, 1)
print(f'Po zmianie:\nWymiary obrazu: {obraz1.size},\nwymiary tła: {obraz4.size}.')
```
