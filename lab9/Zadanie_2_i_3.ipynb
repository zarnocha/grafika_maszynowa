```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
obraz1 = Image.open('obraz1.png')
print(f'Tryb obrazu: {obraz1.mode}')
```

```python
r, g, b, a = obraz1.split()

obraz3 = Image.new('RGB', obraz1.size, (255,255,255))
obraz3.paste(obraz1, (0, 0), a)
print(f'Tryb obrazu: {obraz3.mode}')
```

<h1><center><font size="10">Zadanie 2:</font></center></h1>

<h2><center><font size="6">ppkt b:</font></center></h2>

```python
tryby =  ['CMYK','YCbCr','HSV']

for i, t in enumerate(tryby, start=1):  # Metoda na numerowanie i zarazem pobieranie elementów z listy w pętli
    file_name = "tryb "+ str(t) # Tworzenie tytułów nad obrazami na diagramie
    im_c = obraz3.convert(t)    # Zmiana trybów w pętli

    plt.subplot(1, 3, i)    # Tworzenie miejsca na obrazy
    plt.title(str(file_name))   # Tytułowanie obrazów
    plt.imshow(im_c)    # Wyświetlanie obrazów
    plt.axis('off') # Usuwanie osi z diagramów

    plt.tight_layout()  # "Ściśnięcie" obrazów aby nie były "rozstrzelone"
    plt.savefig('fig1.png')
```

<h2><center><font size="6">ppkt b:</font></center></h2>

```python
tryby =  ['CMYK','YCbCr','HSV']
file_name = "tryb "+ str(tryby[0])
im_c = obraz3.convert(tryby[0])
C, M, Y, K = im_c.split()

plt.subplot(1, 4, 1)
plt.title(f'{file_name}:\nkanał C')
plt.imshow(C, "gray")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title(f'{file_name}:\nkanał M')
plt.imshow(M, "gray")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title(f'{file_name}:\nkanał Y')
plt.imshow(Y, "gray")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title(f'{file_name}:\nkanał K')
plt.imshow(K, "gray")
plt.axis('off')

plt.tight_layout()
plt.savefig('cmyk.png')
```
